```python
import os
import pandas as pd
import shutil
```

```python
base = pd.read_csv('metadata/vggss_test.csv', header=None)
base.head()
```

```python
len(base)
```

```python
base[0].nunique()
```

```python
audios = [s.split('.')[0] for s in os.listdir('VGGSound-test-plus-silent/audio')]
len(audios)
```

```python
frames = [s.split('.')[0] for s in os.listdir('VGGSound-test-plus-silent/frames')]
len(frames)
```

```python
len(set(audios).intersection(base[0].to_list()))
```

```python
len(set(frames).intersection(base[0].to_list()))
```

```python
os.makedirs('audio', exist_ok=True)
os.makedirs('frames', exist_ok=True)
```

```python
for i, row in base.iterrows():
    shutil.copy(os.path.join('VGGSound-test-plus-silent/frames', row[0]) + '.jpg', 'frames/')
    shutil.copy(os.path.join('VGGSound-test-plus-silent/audio', row[0]) + '.wav', 'audio/')
```

```python
extended = pd.read_csv('metadata/vggss_test_plus_silent.csv')
extended.head()
```

```python
len(extended)
```

```python
len(set(audios).intersection(extended['audio'].to_list()))
```

```python
len(set(frames).intersection(extended['video'].to_list()))
```

```python
os.makedirs('extend_audio', exist_ok=True)
os.makedirs('extend_frames', exist_ok=True)
```

```python
for i, (frame, audio, _) in extended.iterrows():
    shutil.copy(os.path.join('VGGSound-test-plus-silent/frames', frame) + '.jpg', 'extend_frames/')
    shutil.copy(os.path.join('VGGSound-test-plus-silent/audio', audio) + '.wav', 'extend_audio/')
```

```python
shutil.rmtree('VGGSound-test-plus-silent')
```
